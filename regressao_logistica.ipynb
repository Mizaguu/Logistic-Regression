```python
import pandas as pd
pd.set_option('display.max_columns', 64)
pd.set_option('display.max_rows', 64)
dados = pd.read_csv('C:/Users/Andre/Documents/DidaticaTech/desodorante.csv')
dados.head()
```

```python
from sklearn.preprocessing import LabelEncoder
def conversao(df):
    le = LabelEncoder()
    for col in df.select_dtypes(include = ['object']):
        df[col] = le.fit_transform(df[col])
    return df
```

```python
conversao(dados)
```

```python
faltantes =(dados.isnull().sum()/len(dados['Product']))
faltantes
```

```python
dados['q8.12'].fillna(dados['q8.12'].mean(),inplace=True)
dados['q8.7'].fillna(dados['q8.7'].mean(),inplace=True)
dados.drop('q8.8', axis=1, inplace = True)
dados.drop('q8.9', axis=1, inplace = True)
dados.drop('q8.10', axis=1, inplace = True)
dados.drop('q8.2', axis=1, inplace = True)
dados.drop('q8.17', axis=1, inplace = True)
dados.drop('q8.18', axis=1, inplace = True)
dados.drop('q8.20', axis=1, inplace = True)
dados.drop('q1_1.personal.opinion.of.this.Deodorant', axis=1, inplace = True)
```

```python
dados.info()
```

```python
faltante =(dados.isnull().sum()/len(dados['Product']))
faltante
```

```python
x = dados.drop('Instant.Liking', axis=1)
y = dados['Instant.Liking']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
sdkd = StratifiedKFold(n_splits=5)
modelo = LogisticRegression(penalty = 'l2', solver='liblinear')
resultado = cross_val_score(modelo,x,y,cv=sdkd)
print(resultado.mean())
```

ajustando parametros

```python
import numpy as np
from sklearn.model_selection import GridSearchCV
valores_c = np.array([0.01,0.1,0.5,0.8,1,2,4,8,10,11,50,90,100])
regularizacao = ['l1','l2']
valores_grid = {'C':valores_c,'penalty':regularizacao}
lr = LogisticRegression(solver='liblinear')
grid_regressao_logistica = GridSearchCV(estimator = lr,param_grid=valores_grid,cv=5)
grid_regressao_logistica.fit(x,y)
print('melhor acuracia',grid_regressao_logistica.best_score_)
```

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
pd.set_option('display.max_columns', 30)
dados = load_breast_cancer()
x = pd.DataFrame(dados.data, columns= [dados.feature_names])
y = pd.Series(dados.target)
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
valores_c2 = np.array([0.01,0.1,0.5,0.8,1,2,4,8,10,11,50,90,100])
regularizacao2 = ['l1','l2']
valores_grid2 = {'C':valores_c2,'penalty':regularizacao2}
lr2 = LogisticRegression(solver='liblinear')
grid_regressao_logistica2 = GridSearchCV(estimator = lr2,param_grid=valores_grid2,cv=5)
grid_regressao_logistica2.fit(x,y)
print('melhor acuracia',grid_regressao_logistica2.best_score_)
```
